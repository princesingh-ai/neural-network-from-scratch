```python
import numpy as np
import matplotlib.pyplot as plt
import keras
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
x_train = x_train.reshape(-1, 28*28) / 255
x_test = x_test.reshape(-1, 28*28) / 255
print(x_train.shape)
print(x_test.shape)
```

```python
w1 = np.random.randn(28*28, 128) * 0.01
b1 = np.zeros((1, 128))
w2 = np.random.randn(128, 10) * 0.01
b2 = np.zeros((1, 10))
```

```python
x = x_train[:10]
label = y_train[:10]
print(x.shape)
print(label)
```

```python
y = np.zeros((10, 10))
y[np.arange(10), label] = 1
```

```python
print(y)
```

```python
for epoch in range(1000):
    # Forward pass
    z1 = np.dot(x, w1) + b1
    a1 = np.maximum(0, z1) # ReLU activation
    z2 = np.dot(a1, w2) + b2
    exp = np.exp(z2 - np.max(z2, axis=1, keepdims=True))
    softmax = exp / np.sum(exp, axis=1, keepdims=True)

    # Loss
    correct_prob = softmax[np.arange(10), label]
    loss = -np.mean(np.log(correct_prob))

    # Bakward pass
    dz2 = softmax - y

    dw2 = np.dot(a1.T, dz2)
    db2 = np.sum(dz2, axis=0, keepdims=True)

    da1 = np.dot(dz2, w2.T)
    dz1 = da1 * (z1 > 0) #Derivative of ReLU

    dw1 = np.dot(x.T, dz1)
    db1 = np.sum(dz1, axis=0, keepdims=True)

    learning_rate = 0.1

    w1 -= learning_rate * dw1
    b1 -= learning_rate * db1

    w2 -= learning_rate * dw2
    b2 -= learning_rate * db2

    if epoch % 100 == 0:
        predicted_class = np.argmax(softmax, axis=1)
        confidence = np.max(softmax, axis=1)
        print(f"Epoch {epoch}: Loss = {loss:.4f}, Predicted class = {predicted_class}, Confidence = {confidence}")
```
